```python
def annualize_rets(r, periods_per_year):
    """
    This function annualizes a set of returns.
    """
    compounded_growth = (1+r).prod()
    n_periods = r.shape[0]
    return compounded_growth**(periods_per_year/n_periods)-1


def annualize_vol(r, periods_per_year):
    """
    This function annualizes the volatility of a set of returns.
    """
    return r.std()*(periods_per_year**0.5)
```

```python
#Functions to be used below.

def portfolio_return(weights, returns):
    """
    This function takes portfolio weights and individual returns as inputs and finds the portfolio expected return.
    """
    return weights.T @ returns


def portfolio_vol(weights, covmat):
    """
    This function takes portfolio weights and the covariance matrix as inputs and finds the portfolio volatility.
    """
    return (weights.T @ covmat @ weights)**0.5


from scipy.optimize import minimize
def minimize_vol(target_return, er, cov):
    """
    This function takes target expected return level, individual returns and covariance matrix as inputs and finds the portfolio with the
    minimum volatility for that particular return level, by applying various constraints.
    """
    n = er.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),)*n #This creates a tuple of n tuples, each having bounds zero and 1
    return_is_target = {
        'type': 'eq', #type of constraint is equality
        'args': (er,), #additional arguments
        'fun': lambda weights, er: target_return - portfolio_return(weights, er) #This is a shorthand way of writing an anonymous function
        #This function will be equal to zero when the initial constraint is met
        #Returns should be equal to whatever we're targeting
    }
    weights_sum_to_1= {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) - 1
        #sum of weights should be equal to 1
    }
    results = minimize(portfolio_vol, init_guess,
                       args=(cov,), method="SLSQP",
                       options={'disp': False},
                       constraints=(return_is_target, weights_sum_to_1),
                       bounds=bounds
                      )
    return results.x #Returns only the weights, not the other results

def optimal_weights(n_points, er, cov):
    """
    This function generates a list of weights to run the optimizer function on; to minimize the volatility for a level of expected return.
    """
    target_rs = np.linspace(er.min(), er.max(), n_points) #Set of target returns between lowest and highest.
    weights = [minimize_vol(target_return, er, cov) for target_return in target_rs]
    return weights


def msr(riskfree_rate, er, cov):
    """
    This function returns the weights of the portfolio that gives you the maximum sharpe ratio given the riskfree rate,
    expected returns and the covariance matrix.
    """
    n = er.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),)*n #This creates a tuple of n tuples, each having bounds zero and 1
    weights_sum_to_1= {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) - 1
        #sum of weights should be equal to 1
    }
    
    def neg_sharpe_ratio(weights, riskfree_rate, er, cov):
        """
        This function returns the negative of the sharpe ratio, given weights, risk free rate, expected returns and the covariance matrix.
        """
        r = portfolio_return(weights, er)
        vol = portfolio_vol(weights, cov)
        return -(r - riskfree_rate)/vol
    
    #The below function minimizes negative sharpe ratio, i.e. maximizes sharpe ratio
    #given various arguments, using the Sequential Least Squares Programming method.
    results = minimize(neg_sharpe_ratio, init_guess,
                       args=(riskfree_rate, er, cov,), method="SLSQP",
                       options={'disp': False},
                       constraints=(weights_sum_to_1),
                       bounds=bounds
                      )
    return results.x #Returns only the optimal weights, not the other results

def gmv(cov):
    """
    This function returns the weights of the Global Minimum Volatility portfolio given the covariance matrix.
    """
    n = cov.shape[0]
    return msr(0, np.repeat(1, n), cov) #In this, you're giving all expected returns to be the same (1, n times), hence maximum sharpe ratio will be the one with minimum variance


def plot_ef(n_points, er, cov, show_cml=False, style='.-', riskfree_rate=0, show_ew=False, show_gmv=False):
    """
    This function plots the efficient frontier for a given expected returns, covariance matrix and risk-free rate.
    """
    weights = optimal_weights(n_points, er, cov)
    rets = [portfolio_return(w, er) for w in weights]
    vols = [portfolio_vol(w, cov) for w in weights]
    ef = pd.DataFrame({
        "Returns": rets,
        "Volatility": vols
    })
    ax = ef.plot.line(x="Volatility", y="Returns", style='.-', title='Markowitz Efficient Frontier')
    if show_ew:
        #This plots the equal-weighted portfolio on the graph.
        n = er.shape[0]
        w_ew = np.repeat(1/n, n)
        r_ew = portfolio_return(w_ew, er)
        vol_ew = portfolio_vol(w_ew, cov)
        sharpe_ew = ()
        print("Equal Weighted Portfolio: Returns=", r_ew, ", Volatility=", vol_ew, ", Annualized Sharpe Ratio=", (r_ew-riskfree_rate)/vol_ew)
        print("------------------------")
        #display EW (Equally weighted portfolio)
        ax.plot([vol_ew], [r_ew], color="goldenrod", marker="o", markersize=12, label='Equally Weighted Portfolio')
        ax.legend(loc='lower left')
              
    if show_gmv:
        #This plots the global minimum variance portfolio on the graph.
        w_gmv = gmv(cov)
        r_gmv = portfolio_return(w_gmv, er)
        vol_gmv = portfolio_vol(w_gmv, cov)
        print("Global Minimum Variance Portfolio: Returns=", r_gmv, ", Volatility=", vol_gmv, ", Annualized Sharpe Ratio=", (r_gmv-riskfree_rate)/vol_gmv)
        print("------------------------")
        # display GMV (Global minimum variance portfolio)
        ax.plot([vol_gmv], [r_gmv], color="midnightblue", marker="o", markersize=12, label='Global Minimum Variance')
        ax.legend(loc='lower left')
        
    if show_cml:
        #This plots the capital market line, which joins the riskless asset and the maximum sharpe ratio portfolio.
        ax.set_xlim(left = 0)
        w_msr = msr(riskfree_rate, er, cov)
        r_msr = portfolio_return(w_msr, er)
        vol_msr = portfolio_vol(w_msr, cov)
        # Add CML
        cml_x = [0, vol_msr]
        cml_y = [riskfree_rate, r_msr]
        print("Maximum Sharpe Portfolio: Returns=", r_msr, ", Volatility=", vol_msr, ", Annualized Sharpe Ratio=", (r_msr-riskfree_rate)/vol_msr)
        print("------------------------")
        ax.plot(cml_x, cml_y, color="green", marker="o", linestyle="dashed", markersize=12, linewidth=2, label='Capital Market Line')
        ax.legend(loc='lower left')
    return ax
```
